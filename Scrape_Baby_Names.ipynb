```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import os
import csv
import sys
import time
```

```python
start = time.time()
connect_url = r'https://parenting.firstcry.com/indian-baby-names/girl-religion-hindu/'
op_location = os.getcwd()
op_file     = 'op_names.csv'

name_list   = op_location + "/" + op_file

if os.path.isfile(name_list):
    response = input("File is present. Do you wish to continue? Existing data will be erased.\nChoose [Y/N] :\n")  
    if response != "Y":
        sys.exit()
else:
    print("Out file is not present. Scraping data from source url :")
    
req_headers = {}
req_headers['user-agent'] = r'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'


try:
    with open(name_list, 'w', encoding='utf-8') as fo:
        fo.write("Name|Meaning|link\n")
        url = connect_url
        print("1" , url)
        while (url):
            # Creating a request object
            req = Request(url, headers=req_headers)

            response = urlopen(req)
            print("THE CONNECTION STATUS = ", response.status)
            page_content = response.read().decode('utf-8')
            bs = BeautifulSoup(page_content,'html5lib')
    #         print("bs done")
    #         htmldata = bs.prettify()
            
            for i in bs.find_all('div', class_="row acadp-entry"):
                meaning = i.find('p', class_="acadp-listings-desc").text.split('-')[1]
                fo.write("{}|{}|{}\n".format(i.a.text , meaning ,i.a['href'] ))
            
            next_url = bs.find('a', class_="next page-numbers")
            if next_url:
                url = next_url['href']
            else:
                print("END OF DATA REACHED.")
                break
            print(url)
    duration = time.time() - start
#     print(duration)
    mins, sec = divmod(duration,60) 
    print("\nDATA EXTRACTION COMPLETED IN {} MINS AND {} SECONDS".format(mins, sec))

except Exception as e:
    print("EXCEPTION BLOCK OF FUNCTION fetchData :", e)
    duration = time.time() - start
    mins, sec = divmod(duration,60) 
    print("\nDATA EXTRACTION INTERRUPTED IN {} MINS AND {} SECONDS".format(mins, sec))
    sys.exit(2)    
```
